```python
import numpy as np
from matplotlib import pyplot as plt
import math
import scipy.integrate as integrate
```

```python
t = np.arange(0, 100, 0.01)

A_t = 2.2*t + 0.085*t**2 - (2/1875)*t**3
D_t = 1.2*t + 0.035*t**2

plt.plot(t, A_t, label = "A(t)")
plt.plot(t, D_t, label = "D(t)")
plt.legend()
plt.xlabel('Temps (minutes)')
plt.ylabel("Nombre cumulé de véhicules")
plt.title('Nombre de véhicules arrivés et servis')
# plt.savefig('Donnees_produites/Figures/arrivees_depart.eps', format='eps')
plt.show()
```

```python
# Exercice 1
A_t = lambda t: 2.2*t + 0.085*t**2 - (2/1875)*t**3
D_t = lambda t: 1.2*t + 0.035*t**2

# Point d'intersection entre les courbes
intersec_x = 60

i1 = integrate.quad(A_t, 0, intersec_x)[0]
i2 = integrate.quad(D_t, 0, intersec_x)[0]

Aire = i1 - i2

# Retard total
D_tot = Aire
print("%d minutes"%D_tot)
```
